```python
import pandas as pd
import geopandas as gpd
import json
```

```python
# abrimos el archivo JSON
with open('CAP.json') as f_in:
       data = json.load(f_in)
```

```python
data
```

```python
# recorremos el campo de longitud redondeando el valor
for i in range(0,len(data["features"])):
    data["features"][i]['attributes']['Longitud'] = round(data["features"][i]['attributes']['Longitud'], 2)
```

```python
# guardar el archivo JSON
with open('data.json', 'w') as f:
    json.dump(data, f)
```
